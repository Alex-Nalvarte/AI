# Is there a relationship between the number of reviews and the price of a book on the bestseller List?

This is an observational study because I am not changing any variables but just looking at existing data. This dataset contains books that appeared on Amazon's Bestseller List and is a sample of books rather than the whole population of published books. A potential bias is that since this data only includes bestselling books, it can make pricing and ratings seem higher than the average book because of their popularity and success.  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('books.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Reviews', y='Price', data=df, scatter_kws={'alpha':0.5})
plt.scatter(df['Reviews'], df['Price'], alpha=0.6)
plt.title('Book Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price (USD)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Price'], bins=20, kde=True)
plt.title('Distribution of Book Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Count')
plt.show()
```

```python
mean_price = df['Price'].mean()
mean_reviews = df['Reviews'].mean()
median_price = df['Price'].median()
median_reviews = df['Reviews'].median()
mode_price = df['Price'].mode()
mode_reviews = df['Reviews'].mode()
```

```python
print("---- Price Statistics ----")
print(f"Mean Price: ${mean_price:.2f}")
print(f"Median Price: ${median_price:.2f}")
print(f"Mode Price: {', '.join(map(str, mode_price.values))}")

print("\n---- Reviews Statistics ----")
print(f"Mean Reviews: {mean_reviews:.0f}")
print(f"Median Reviews: {median_reviews}")
print(f"Mode Reviews: {', '.join(map(str, mode_reviews.values))}")
```

```python
r, p = pearsonr(df['Reviews'], df['Price'])
print(f"Pearson r: {r:.3f}")
```

There is no meaningful relationship between the number of reviews and book price in the dataset analyzed. The Correlation Coefficient was -0.109 which indicates a weak negative relationship between the two variables. The Scatterplot with regression line helps visualize this relationship.
