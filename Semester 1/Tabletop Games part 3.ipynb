# In Monopoly, what is the probability that you roll a 5 with one Dice and an even number with the other dice?

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
trials = 10000
```

```python
results = []

for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    
    condition_met = (die1 == 5 and die2 % 2 == 0) or (die2 == 5 and die1 % 2 == 0)
    results.append(condition_met)
```

```python
success_rate = sum(results) / trials
print(f"Proportion of trials where one die was 5 and the other even: {success_rate:.2f}")
```

```python
sns.countplot(x=results)
plt.xticks([0, 1], ['Condition Not Met', 'Condition Met'])
plt.title('Simulation Results: One Die is 5 and Other is Even')
plt.ylabel('Number of Trials')
plt.show()
```

The Theoretical probability that you roll a 5 with one dice and an even number with another dice in Monopoly is 1 in 6. (16.7 percent). The probability of rolling a 5 on one dice is 1/6, and the probability of rolling an even number on one dice is 3/6 (1/2). Multiplying 1/6 and 1/2 gets us 1/12. If you add them you get 2/12, which simplifies to 1 in 6. The empirical probability obtained from my simulation is 17 percent (Really close to 16.7 percent), I ran 10,000 trials and about 1,700 of them met both conditions.
