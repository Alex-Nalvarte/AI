```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def roll_die(sides=6, trials=100):
    results = []
    for _ in range(trials):
        roll = random.randint(1, sides)
        results.append(roll)
    return results 
```

```python
results = roll_die()
print(results[:10])
```

```python
sns.histplot(results, bins=range(1, 8), discrete=True, kde=False)
plt.title('Distribution of 100 Rolls of a 20-sided Die')
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.show()
```
