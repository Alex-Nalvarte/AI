# In Uno, what is the probability that the first card dealt is either a 0 or a yellow card? (Assume no other cards have been dealt)

I will determine the probability that the first card is either a 0 or a yellow card in Uno (assuming no other cards have been dealt) with a simulation. The probability of drawing a zero is 4/128, and the probability of drawing a yellow card is 25/128. The probability that just one of these happens is 1 in 128. 

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def create_uno_deck():
    colors = ['Red', 'Green', 'Blue', 'Yellow']
    numbers = list(range(10))
    actions = ['Skip', 'Reverse', 'Draw Two']
    deck = []
    for color in colors:
        deck.append((color, 0))  # one zero
        for num in range(1, 10):
            deck.extend([(color, num)] * 2)
        for action in actions:
            deck.extend([(color, action)] * 2)
    wilds = ['Wild', 'Wild Draw Four']
    for wild in wilds:
        deck.extend([('Black', wild)] * 4)
    return deck
```

```python
def is_success(deck):
    first_card = random.choice(deck)
    color, value = first_card
    return (value == 0) or (color == 'Yellow')
```

```python
n_trials = 10000
results = [is_success(create_uno_deck()) for _ in range(n_trials)]
```

```python
counts = {'Success (0 or Yellow)': results.count(True), 'Failure': results.count(False)}
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(6,4))
sns.barplot(x=list(counts.keys()), y=list(counts.values()), palette='pastel')
plt.ylabel('Count')
plt.title(f'Results of {n_trials} Uno Card Draws')
plt.show()
```

The probability that either a 0 or yellow card is dealt first is 25 percent. After running 10,000 trials, about 2500 of them either had a 0 or a yellow. (Empircial probability) The theoretical probability is also about 25 percent. 4/108 + 25/102 - 1/108 = 28/108 which simplifies to 7/27. 7 Divided by 27 is 0.259 or 25.9 percent.
